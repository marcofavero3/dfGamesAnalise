```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings

warnings.filterwarnings('ignore')

url = 'https://raw.githubusercontent.com/marcofavero3/dfGamesAnalise/main/PS4_GamesSales.csv'

df = pd.read_csv(url, encoding='ISO-8859-1') #necessário adicionar o encoding para aceitar a leitura, visto que não possui a formatação correta

df.head()
```

```python
#analisar a dimensão

df.shape
```

```python
#verificar os nulos

df.isnull().sum()
```

```python
#grafico dos nulos

plt.figure(figsize=(14,6))
plt.title('Verificando campos nulos')

sns.heatmap(df.isnull(), cbar=False);
```

```python
#Remover valores nulos

df.dropna( inplace=True) #comando inplace ele sobrescreve o DF
```

```python
#Ver estatisticas
df.describe()
```

```python
sns.barplot(data=df, x='Year', y='Global', ci=None, estimator=sum)

# Adiciona o título do gráfico
plt.title('Vendas Globais em Milhões', loc='left', fontsize=14)

# Adiciona o rótulo do eixo y
plt.ylabel('Qnt de vendas')

# Exibe o gráfico
plt.show()
```

```python
#Remover anos 2019 e 2020 e mostrar os anos que me "restaram"

df = df.loc[(df['Year'] !=2019) & (df['Year'] !=2020)] #['Year'].unique()

df.head()
```

```python
#Plotar a distruição do valor GLOBAL

#TAMANHO
plt.figure(figsize=(12,5))

#ESTILO
plt.style.use('ggplot')

#PLOTAGEM
plt.title('Distribuição das vendas Globais', loc='left', fontsize=14) #titulo
sns.kdeplot(df['Global'], shade=True, bw=1, linewidth=2.5) #shade e bw são comandos para passar a transparencia/sombreamento da curva do gráfico
```

```python
#Verificar a distribuição por ano/local
df.groupby(by=['Year']).sum()
```

```python
#PLOTAGEM
plt.figure(figsize=(10,4))
plt.title('Distribuição Global')

sns.boxplot(data=df, x='Year', y='Global');
```

```python
#Criando uma base de corte de jogos com mais de 10 milhões de vendas

df.loc[df['Global'] >=10]
```

```python
Analise = df.groupby( by=['Year'] ).sum().reset_index()

# analisando a proporção dos 100% de cada continemente comparado ao Total
America = [ America / Total * 100 for America, Total in zip( Analise['North America'], Analise['Global'] ) ]
Europa = [ Europa / Total * 100 for Europa, Total in zip( Analise['Europe'], Analise['Global'] ) ]
Japao = [ Japao / Total * 100 for Japao, Total in zip( Analise['Japan'], Analise['Global'] ) ]
Mundo = [ Mundo / Total * 100 for Mundo, Total in zip( Analise['Rest of World'], Analise['Global'] ) ]

America, Europa, Japao, Mundo
```

```python
#Plotar grafico de barras empilhado

plt.figure(figsize=(10,5))

#config do grafico
Largura_Barra = 0.85
Rotulos = Analise['Year']
Grupos = [0,1,2,3,4,5]
#titulo
plt.title('Analise da distribuição por continentes')

#plot america
plt.bar( Grupos, America, width=Largura_Barra, color='#b5ffb9', edgecolor='white') #VERDE
#plot europa
plt.bar( Grupos, Europa, bottom=America, width=Largura_Barra, color='#f9bc86', edgecolor='white') #LARANJA
#plot japao
plt.bar( Grupos, Japao, bottom=[A+B for A,B in zip(America,Europa)], width=Largura_Barra, color='#a3acff', edgecolor='white')#AZUL
#plot resto do mundo
plt.bar( Grupos, Mundo, bottom=[A+B+C for A,B,C in zip(America,Europa,Japao)], width=Largura_Barra, color='#d3acfe', edgecolor='white')

#Labels
plt.xticks(Grupos, Rotulos)
plt.xlabel('Grupo')
plt.ylabel('Distribuição %')

#Legenda
plt.legend(['America Norte','Europa','Japao','Mundo'], loc='upper left', bbox_to_anchor=(0.15,-0.1), ncol=4);

```

```python
#CONVERSÃO DE VALORES PARA NÚMEROS, E ATRIBUIÇÃO DE COLUNAS, FIZ A ALTERAÇÃO DE VALORES NUMERICOS TROCANDO PELOS NOMES, DESSA FORMA FICANDO MAIS LEGIVEL
from sklearn.preprocessing import LabelEncoder
Funcao_Label = LabelEncoder()

df['Produtor'] = Funcao_Label.fit_transform(df['Publisher'])
df['Genero'] = Funcao_Label.fit_transform(df['Genre'])
df['Jogo'] = Funcao_Label.fit_transform(df['Game'])

df.head()
```

```python
#Paleta de Cores
Paleta_Cores = sns.color_palette('husl',8)
Paleta_Cores
```

```python
#PLOTAGEM DAS PRODUTORAS DE GAMES
plt.figure(figsize=(20,5))
plt.title('Analise dos Produtores de Games')

sns.scatterplot(data=df, x='Produtor', y='Global', color=Paleta_Cores[0])
```

```python
#PLOTAGEM DO GENERO DOS GAMES

plt.figure(figsize=(20,5))
plt.title('Analise dos Generos dos Games')

sns.scatterplot(data=df, x='Genero', y='Global', color=Paleta_Cores[1])
```

```python
# PLOTAGEM DOS GAMES


plt.figure(figsize=(20,5))
plt.title('Analise dos Games')

sns.scatterplot(data=df, x='Jogo', y='Global', color=Paleta_Cores[2])
```
